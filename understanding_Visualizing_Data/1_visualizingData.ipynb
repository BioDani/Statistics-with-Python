# Visualizing data in Python

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
data.total_bill.describe()
```

```python
# Histograma de total_bill
sns.displot(data["total_bill"], kde = False)
plt.show()
```

```python
# Histograma de tip
sns.displot(data["tip"], kde = False)
plt.show()
```

```python
# Create a boxplot and histogram of the tips grouped by smoking status
sns.boxplot(x = data["smoker"], y = data["tip"])
plt.show()
```

```python
# Create a boxplot and histogram of the tips grouped by time of day
sns.boxplot(x = data["time"], y = data["tip"])
```

```python
sns.boxplot(x = data["day"], y = data["tip"])
```
